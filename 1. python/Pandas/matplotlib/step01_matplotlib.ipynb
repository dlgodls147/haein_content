### 플롯에 타이틀과 레이블 추가
- 아직 한글에 대한 인코딩이 되어있지 않아 한글 작성 불가

```python
plt.plot(p,p+0, ls='-')
plt.plot(p,p+1, ls=':')
plt.plot(p,p+2, ls='--')
plt.plot(p,p+3, ls='-.')

plt.title('DATA VIEW',fontsize=20)
plt.xlabel('x data')
plt.ylabel('y data')
```

```python
plt.plot([1,2,3],[1,4,9], 'bs', label='line 2' , ls = '-')

# 주의 사항 : 여러개의 플랫을 덩시 작업시 범례(라벨표기)
# 구분을 위한 필수 함수
plt.legend()
```

```python
plt.plot([1,2,3],[1,4,9], 'bs', label='line 2' , ls = '-')
plt.plot([2,3,4],[2,5,10], 'gs', label='line 3' , ls = ':')
plt.legend()
```

```python
plt.plot([1,2,3],[1,4,9], 'bs', label='line 2' ,ls='-')
plt.plot([2,3,4],[2,5,10], 'gs', label='line 3' ,ls=':')
# 축 상의 위치 표시 지점을 의미
plt.xticks([0,5,10])
plt.yticks([0,10,20])
plt.legend()
```

# 산점도(scatter plot)
1. 라인 플롯과 흡사
2. 산점도에서 점은 선으로 연결되는 대신 개별적인 dot으로 원 또는 다른 모양으로 표현됨
3. 데이터를 x축과 y축에 점으로 표현한 그래프
4. 나이와 소득처럼 연속된 두 변수의 관계를 표현할 때 사용
5. 데이터의 탐색돠 표현에 적합

```python
# 난수발생한 1번째 값을 고정시킴
data = np.random.RandomState(0)
data.rand(11)
```

```python
'''
제시된 마커 표기로 list 생성
반복문으로 list 내에 있는 마커를 표기
'''
data = np.random.RandomState(0)
marker = ['o','.',',','x','+','v','^','<','>','s','d']
for i in marker:
    plt.plot(data.rand(5),data.rand(5), i ,label="marker={0}".format(i) )

plt.legend(loc=1) # plt .legend? loc=1 생략해도 동일한 위치해 그려짐
plt.xlim(0,1.5)
```

## plot와 scatter의 차이점
- scatter는 각 점의 속성(크기,색상 등) 개별적으로 제어가 가능
- plot는 속성이 동일

```python
plt.plot(x,y,marker='o')
```

```python
plt.scatter(x,y,marker='o') # 산점도 함수 scatter()
```

```python
rng = np.random.RandomState(0) # 이 설정으로 인해 test 데이터를 동일하게 처리 가능
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

#alpha=0.3  겹치는 효과를 위한 투명도
#cmap='viridis' : 생략시 흑색
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar();  # 색상 척도 표시
```

```python
#rng = np.random.RandomState(0) # 이 설정으로 인해 test 데이터를 동일하게 처리 가능
x = rng.randn(100) # 정규분포 안에서의 랜덤값
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

#alpha=0.3  겹치는 효과를 위한 투명도
#cmap='viridis' : 생략시 흑색
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,cmap='viridis')
plt.colorbar();  # 색상 척도 표시
```
